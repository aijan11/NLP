## TOKENIZATION

Tokenization in NLP is the process of breaking down text into smaller units called tokens. 
These tokens can be words, characters or subwordsand sentences depending on the method.

```python
#Tokenization Example
corpus = """Natural language processing is a fascinating field of artificial intelligence.
It enables computers to understand and generate human language!.
NLP has applications in chatbots, sentiment analysis, and machine translation."""
```

```python
print(corpus)
```

```python
# Punctuation package 
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import sent_tokenize
sentences = sent_tokenize(corpus)
sentences
```

```python
#to iterate through sentences
for i in sentences:
    print(i)
```

```python
# convert sentences into words
from nltk.tokenize import word_tokenize
words = word_tokenize(corpus)
words
```

```python
# using loop to for word tokenize
for i in sentences:
    print(word_tokenize(i))
```

```python
# using wordpunct_tokenize to split words and punctuation
from nltk.tokenize import wordpunct_tokenize
wordpunct = wordpunct_tokenize(corpus)
wordpunct
```

```python
# using TreebankWordTokenizer to not tak
from nltk.tokenize import TreebankWordTokenizer
treebank = TreebankWordTokenizer()
treebank.tokenize(corpus)
```

# STEMMING
>> It is the process of reducing a word to its word stem that affixes to suffixes and prefixes or
>> to the roots of words known as lemma. Stemming is important in NLU nad NLU

```python
words = ['eating','writing','swimming','running','writes','swims','runs','ate','wrote','swam','ran'
         ,'quick','quickly','happily','happier','happiest']
```

# PORTER STEMMER

```python
from nltk.stem import PorterStemmer
stemming = PorterStemmer()
for word in words:
    print(word + "--->" + stemming.stem(word))
```

```python
#Disadvantages of Stemming
#1. Stemming can lead to loss of meaning and context, 
# as it reduces words to their base form without considering their specific usage or nuances.
#2. Stemming can produce non-standard or incorrect word forms,
#  which may not be easily recognizable or understandable to users. 
#EXAMPLE: The word "Congratulations" may be stemmed to "Congratuli", which is not a valid word in English and can lead to confusion.
stemming.stem("Congratulations")
```

# RegexpStemmer class

Regexp stemmer basically takes regular singular expression and removes any prefix or suffix that matches the expression

```python
from nltk.stem import RegexpStemmer
regexp = RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
regexp.stem("eating")
```

```python
words = ['eating','writing','swimming','running','writes','swims','runs','ate','wrote','swam','ran']
for word in words:
    print(word + "--->" + regexp.stem(word))
```

# SnowBall Stemmer

Snowball Stemmer is an improvement over the Porter Stemmer, as it provides better handling of irregular word forms and is more effective in reducing words to their base form.Snowball stemmer is provided with different languages.

```python
from nltk.stem import SnowballStemmer
snowball = SnowballStemmer('english')
for word in words:
    print(word + "--->" + snowball.stem(word))
```

# Difference between Snowball stemmer and Porter stemmer

```python
stemming.stem("fairly"),stemming.stem("sportingly") # porter stemming
```

```python
snowball.stem("fairly"),snowball.stem("sportingly") # snowball stemming
```
